# Preprocess the Text

```python
import pandas as pd
from bs4 import BeautifulSoup
import unicodedata
import re
import nltk
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('wordnet')
```

```python
# Fill NaN values in 'reviewText' column with an empty string
merged_df['reviewText'] = merged_df['reviewText'].fillna('')

# Define a function to remove HTML tags from a string
def remove_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)

# Define a function to remove accented characters from a string
def remove_accented_chars(text):
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    return text

# Define a function for expanding acronyms (You can define your own dictionary of acronyms)
def expand_acronyms(text):
    # Define a dictionary of acronyms and their expansions
    acronyms = {
        'lol': 'laugh out loud',
        'brb': 'be right back',
        # Add more acronyms and expansions as needed
    }
    # Replace acronyms with their expansions
    for acronym, expansion in acronyms.items():
        text = text.replace(acronym, expansion)
    return text

# Define a function to remove special characters from a string
def remove_special_characters(text):
    pattern = r'[^a-zA-Z0-9\s]'
    return re.sub(pattern, '', text)

# Define a function for lemmatization
def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(text)
    lemmatized_text = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(lemmatized_text)

# Define the text normalizer function
def normalize_text(text):
    text = remove_html_tags(text)
    text = remove_accented_chars(text)
    text = expand_acronyms(text)
    text = remove_special_characters(text)
    text = lemmatize_text(text)
    return text

# Apply the text normalizer function to the 'reviewText' column
merged_df['normalize_reviewText'] = merged_df['reviewText'].apply(normalize_text)
```

```python
# Define a function to categorize ratings
def categorize_rating(rating):
    if rating > 3:
        return 1
    elif rating == 3:
        return 0
    else:
        return -1

# Apply the function to create a new column 'Rating_Class'
hashing_df['Rating_Class'] = merged_df['overall'].apply(categorize_rating)
```

```python
type(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from tqdm.notebook import tqdm
classifiers = {
    'Random Forest': RandomForestClassifier(n_jobs=-1),
    'Gradient Boosting': GradientBoostingClassifier(),
    'Naive Bayes': GaussianNB(),
    'Logistic Regression': LogisticRegression(n_jobs=-1),
    'Support Vector Machine': SVC()
}

classification_reports = {}
for name, classifier in tqdm(classifiers.items(), total=len(classifiers), desc="Training and Evaluating"):
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    report = classification_report(y_test, y_pred, target_names=[1, 0, -1], digits=4, output_dict=True)
    classification_reports[name] = report
```
